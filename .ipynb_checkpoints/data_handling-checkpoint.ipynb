```python
import pandas as pd
import time 
import json
import urllib.request
from datetime import timedelta
import matplotlib.pyplot as plt
import numpy as np#
from pandas.io.json import json_normalize
```

Importing RPi Data 

```python
df = pd.read_json("https://alte-rs.ddnss.de/weather/processed/history.json")
```

```python
df['time'] = pd.to_datetime(df['time'])
```

```python
df['temp'] = pd.to_numeric(df['temp'])
```

```python
df['pressure'] = pd.to_numeric(df['pressure'])
```

```python
df['humidity'] = pd.to_numeric(df['humidity'])
```

```python
dataframe_rpi = df.set_index('time')
```

```python
rolling = dataframe_rpi.rolling(window='120min',min_periods=10).mean()
```

```python
rolling.isnull().sum()
```

Importing OWM Data

```python
with urllib.request.urlopen("https://alte-rs.ddnss.de/weather/processed/history_owm.json") as url:
    data2 = json.loads(url.read().decode())
```

```python
temp = json_normalize(data2)
```

```python
temp['dt'] = pd.to_datetime(temp['dt'],unit='s')
```

```python
temp['main.temp'] = pd.to_numeric(temp['main.temp'])-273.15
```

```python
temp['main.humidity'] = pd.to_numeric(temp['main.humidity'])
```

```python
temp['main.pressure'] = pd.to_numeric(temp['main.pressure'])
```

```python
dataframe_owm = temp.set_index('dt')
```

```python
dataframe_owm.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(rolling['temp']-15,'b-',dataframe_owm['main.temp'],'r-')
```
